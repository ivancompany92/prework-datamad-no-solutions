# Rock Paper Scissors

Let's play the famous game against our computer.

```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
import random 

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
list = ['stone', 'paper', 'scissors']

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_games=7

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
games_win=3

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

def machine():
    
    machine_1=str(random.choice(list))
    return machine_1


# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def player():
    player_1=input(str('Introduzca una opcion: stone/paper/scissors= '))
    while (player_1!='stone')&(player_1!='paper')&(player_1!='scissors'):  
        player_1=input(str('Introduzca una opcion: stone/paper/scissors= '))
        
    return player_1

    
# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def combat(a,b):  #a=maquina, b=jugador
    if a==b:
        return 0
    elif ((a=='stone')&(b=='paper'))or((a=='paper')&(b=='scissors'))or((a=='scissors')&(b=='stone')):
        return 2
    else:
        return 1
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def game(d,e,f):
    print('El jugador ha elegido: ',e)
    print('La máquina ha elegido: ',d)
    
    if f==0:
        return print('empate')
    elif f==1:
        return print('la maquina gana esta partida')
    elif f==2:
        return print('el jugador gana esta partida')


    
# Create two variables that accumulate the wins of each participant
player_wins=0
machine_wins=0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
games=0

while (player_wins!=games_win)&(machine_wins!=games_win)&(games!=max_games):
    a=machine()
    b=player()
    c=combat(a,b)
    if c==1:
        machine_wins+=1
    elif c==2:
        player_wins+=1
    
    game(a,b,c)
    games+=1
    
    
# Print by console the winner of the game based on who has more accumulated wins
if player_wins==games_win:
    print('El JUGADOR gana el JUEGO')
elif machine_wins==games_win:
    print('La MAQUINA gana el JUEGO')
else:
    print('El JUEGO termina sin que el jugador o la maquina hayan ganado conseguido las ',games_win,'partidas necesarias para ganar el juego')
```

```python
# Import the choice function of the random module
import random 


# Define a function that asks for an odd number on the keyboard, until it is not valid
# will keep asking
def impar():
    while True:
        try:
            ab=float(input('Introduzca un numero impar: '))
            if ((ab//2)!=(ab/2)):
                break
            else:
                print('No introduzco un numero impar.')
        except ValueError:
            print('No introduzco un numero impar.')
      
    return int(ab)


# Assign a list of 5 possible options.
list_2 = ['stone', 'paper', 'scissors','lizard', 'spock']


# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
# This time the previously defined function is used
max_games_2=impar()

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
games_win_2=(max_games_2//2)+1

# Define a function that randomly returns one of the 5 options.
# This will correspond to the play of the machine. Totally random.

def machine_2():
    
    machine_1=str(random.choice(list_2))
    return machine_1


# Define a function that asks your choice between 5
# you should only allow one of the 5 options. This is defensive programming.
# If it is not valid, keep asking until it is valid.
def player_2():
    play=input(str('Introduzca una opcion: stone/paper/scissors/lizard/spock= '))
    while (play!='stone')&(play!='paper')&(play!='scissors')&(play!='lizard')&(play!='spock'):  
        play=input(str('Introduzca una opcion: stone/paper/scissors/lizard/spock= '))
        
    return play


# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
# Now there are more options
def combat_2(a,b): #a=maquina, b=jugador
    if a==b:
        return 0
    elif ((a=='stone')&((b=='paper')or(b=='spock')))or((a=='paper')&((b=='scissors')or(b=='lizard')))or((a=='scissors')&((b=='stone')or(b=='spock')))or((a=='lizard')&((b=='scissors')or(b=='stone')))or((a=='spock')&((b=='paper')or(b=='lizard'))):
        return 2  #formas en la que ganaria el jugador a la maquina.
    else:
        return 1


    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def game_2(d,e,f):
    
    print('La máquina ha elegido: ',d)
    print('El jugador ha elegido: ',e)
    
    if f==0:
        return print('empate')
    elif f==1:
        return print('la maquina gana esta partida')
    elif f==2:
        return print('el jugador gana esta partida')


    
# Create two variables that accumulate the wins of each participant
player_wins_2=0
machine_wins_2=0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
games=0

while (player_wins_2!=games_win_2)&(machine_wins_2!=games_win_2)&(games!=max_games_2):
    a=machine_2()
    b=player_2()
    c=combat_2(a,b)
    if c==1:
        machine_wins_2+=1
    elif c==2:
        player_wins_2+=1
    
    game_2(a,b,c)
    games+=1
    
    
# Print by console the winner of the game based on who has more accumulated wins
if player_wins_2==games_win_2:
    print('El JUGADOR gana el JUEGO')
elif machine_wins_2==games_win_2:
    print('La MAQUINA gana el JUEGO')
else:
    print('El juego termina sin que el jugador o la maquina hayan ganado conseguido las ',games_win_2,'partidas necesarias para ganar el juego')
```
